```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://insights.blackcoffer.com/how-does-marketing-influence-businesses-and-consumers/'
req = requests.get(url)
html_page = req.content
html_page
```

```python
soup = BeautifulSoup(html_page,'html.parser')
soup
```

```python
Title = soup.find('title')
Title
```

```python
Clean_Title = Title.get_text()
print(Clean_Title)
```

```python
Article = soup.find('div',attrs = {'class' : "td-post-content tagdiv-type"})
Article
```

```python
Clean_Article = Article.get_text()
print(Clean_Title , Clean_Article)
```
